4. **Hur skiljer sig IoT-data från traditionell affärsdata och vilka konsekvenser får detta för BI-arbete?**
**Traditionell data** - försäljningsdata, transaktionsdata, lagerstatus, den registreras när någon gör någonting aktivt, från en människa eller ett system. Ex man matar in en försäljning. Data kommer från när någotning händer, den är historisk, när ngåotning hänt historiskt, analys av data görs i efterhand, vi samlar mycket data innan vi kan göra analys. 
**IoT-data** - data som skapats av uppkopplade fysiska enheter.  Extremt mycket data, större krav på hur man ska hantera data, hur man ska spara data, rensa data osv. 
Vi kan med IoT-data reagera i realtid, medan traditionell data är i efterhand eftersom den är historisk.
